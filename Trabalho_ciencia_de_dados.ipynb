```python
import plotly.graph_objects as go

dados_mulheres = dados[dados['Gender'] == 'F']

dados_homens = dados[dados['Gender'] == 'M']

media_mulheres = dados_mulheres.groupby('Question')['Value'].mean().reset_index()
media_homens = dados_homens.groupby('Question')['Value'].mean().reset_index()

media_mulheres = media_mulheres.sort_values(by='Value', ascending=False)
media_homens = media_homens.sort_values(by='Value', ascending=False)

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=media_mulheres['Value'],
    theta=media_mulheres['Question'],
    fill='toself',
    name='Mulheres'
))

fig.add_trace(go.Scatterpolar(
    r=media_homens['Value'],
    theta=media_homens['Question'],
    fill='toself',
    name='Homens'
))

fig.update_layout(
    polar=dict(
        radialaxis=dict(visible=True, range=[0, max(media_mulheres['Value'].max(), media_homens['Value'].max())])
    ),
    showlegend=True,
    title='Média de Casos por Tipo de Violência e Gênero (Radar)'
)

fig.show()
#pyo.plot(fig, filename='tipos de violencia3')
```

```python
import plotly.graph_objects as go

categorias_emprego = ['Employed for income', 'Employed for household duties (e.g., children, elderly)', 'Unemployed', 'Employed for cash', 'Employed for kind']
dados_emprego = dados[dados['Demographics Response'].isin(categorias_emprego) & (dados['Gender'] == 'F')]

media_por_emprego = dados_emprego.groupby('Demographics Response')['Value'].mean().reset_index()

categories = media_por_emprego['Demographics Response']
values = media_por_emprego['Value']

fig_radar = go.Figure()

fig_radar.add_trace(go.Scatterpolar(
      r=values.tolist() + [values.tolist()[0]],
      theta=categories.tolist() + [categories.tolist()[0]],
      fill='toself',
      name='Média de Casos',
      line_color='purple'
))

fig_radar.update_layout(
    polar=dict(
        radialaxis=dict(visible=True, range=[0, max(media_por_emprego['Value']) + 1.5])
    ),
    showlegend=True,
    title='Violência contra Mulheres por Situação de Emprego (Média)'
)

for cat, val in zip(categories, values):
    fig_radar.add_trace(go.Scatterpolar(
        r=[val],
        theta=[cat],
        mode='markers',
        marker=dict(color='red', size=10),
        showlegend=False
    ))
#py.plot(fig_radar, filename='violencia-mulheres-emprego-radar', auto_open=False)
fig_radar.show()
```

```python

import plotly.subplots as sp
import plotly.graph_objects as go


dados['Survey Year'] = pd.to_datetime(dados['Survey Year'], errors='coerce')

dados_localizacao = dados[dados['Demographics Response'].isin(['Rural', 'Urban'])]

media_geral = dados_localizacao.groupby(['Gender', 'Demographics Response'])['Value'].mean().reset_index()

media_mulheres = media_geral[media_geral['Gender'] == 'F']
media_homens = media_geral[media_geral['Gender'] == 'M']

cores = {'Rural': 'orange', 'Urban': 'blue'}

fig = sp.make_subplots(rows=1, cols=2, subplot_titles=['Mulheres', 'Homens'], specs=[[{'type':'domain'}, {'type':'domain'}]])


fig.add_trace(go.Pie(labels=media_mulheres['Demographics Response'],
                     values=media_mulheres['Value'],
                     marker_colors=[cores[d] for d in media_mulheres['Demographics Response']],
                     hole=0.3, name=''),
              row=1, col=1)

fig.add_trace(go.Pie(labels=media_homens['Demographics Response'],
                     values=media_homens['Value'],
                     marker_colors=[cores[d] for d in media_homens['Demographics Response']],
                     hole=0.3, name=''),
              row=1, col=2)

fig.update_layout(title_text='Média Geral de Violência por Localização')

#py.plot(fig, filename='media-violencia-localizacao', auto_open=False)
fig.show()
```
